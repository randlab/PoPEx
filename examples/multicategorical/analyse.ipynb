```python
import numpy as np
import matplotlib.pyplot as plt

import pickle
from popex import algorithm, utils
from popex.popex_objects import Problem, CatParam, PoPEx
```

```python
def get_pickle(name):
    result = None
    try:
        with open(name, 'rb') as file_handle:
            result = pickle.load(file_handle)
    except FileNotFoundError as e:
        print(f'{name} failed: {str(e)}')
    return result
```

```python
NW_MIN = 100
```

```python
with open(f'popex/popex.pop', 'rb') as file_handle:
    popex_sim = pickle.load(file_handle)
```

```python
weights = utils.compute_w_pred(popex_sim, nw_min=NW_MIN)
catprob = utils.compute_cat_prob(popex_sim, weights=weights)
plt.imshow(catprob[0].param_val[:,1].reshape(128,128), cmap='magma', origin='lower')
plt.colorbar()
```
